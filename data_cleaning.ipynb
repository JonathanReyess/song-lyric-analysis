```python
def clean_lyrics(df,  decade, genre = None, language = 'en'):

  # df: pandas dataframe, this is ALL of the lyrics data
  # genre: (str), what genre we want to filter by
  # decade: (str), what decade the music is from (1 == 2010-2019, 4 = 1940-49, etc)
  # language: (str), what language we want lyrics from


  if genre != None:
    song_df = df[(df['playlist_genre'] == genre)]

  else:
    song_df = df
    genre = 'ALL-GENRES'

  song_df = song_df[song_df['language'] == language]
  song_df = song_df[song_df['track_album_release_date'].apply(lambda x: x.split('-')[0][2]) == decade]

  outDF = song_df['lyrics']

  name = f"{language}_{genre}__{decade}0's_Lyrics.csv"

  outDF.to_csv(name)

  print(f"Writing file {name}")
```

```python
genres = list(df['playlist_genre'].unique())
genres
```

```python
decades = list((range(0,10)))
decades = [str(x) for x in decades]
decades
```

```python
for decade in decades:
  clean_lyrics(df, decade=decade)
print("Job's Done!")
```

```python
for genre in genres:
  clean_lyrics(df, genre=genre, decade = '9')

print("Job's Done!")
```
